> Question3: Compute margin of error, confidence interval, and p-value.

The code and result is below.

```python
# Find out the total number, the call back number, and the no call back number for black and white applicants
bl_total = len(data[data.race=='b'])
wh_total = len(data[data.race=='w'])
bl_callb = len(data[(data.race=='b') & (data.call==1)])
wh_callb = len(data[(data.race=='w') & (data.call==1)])
bl_nocall = bl_total - bl_callb
wh_nocall = wh_total - wh_callb
print(bl_callb,wh_callb,bl_nocall,wh_nocall)
```

```python
# Find out the expected call back ratio(expected_cr), and build the observed matrix and the expected matrix
expected_cr =( bl_callb + wh_callb)/(bl_total + wh_total)
df = pd.DataFrame({'Black':[bl_nocall, bl_callb], 'White':[wh_nocall, wh_callb]})
df_expected = pd.DataFrame({'Black':[bl_total * (1 - expected_cr), bl_total*expected_cr], 'White': [wh_total*(1-expected_cr), wh_total*expected_cr]})
print(df)
print(df_expected)
```

```python
# Calculate the chi square test statistic
df1 = (df-df_expected)**2/df_expected 
print(df1)
chi_value = df1.sum().sum()
print('chi value:', chi_value)
```

The degree of freedom is (2-1)x(2-1)=1, and the corresponding p value is smaller than 0.5%, which means that the Null hypothesis is rejected, in other words, the race and the call back ratio is not independent, the racial discrimination does exist.
